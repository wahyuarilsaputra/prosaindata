# Scrapping Data dari komentar video Youtube 

```python
import requests
from bs4 import BeautifulSoup
```

```python
def crawl_comments(video_id):
    url = f'https://www.youtube.com/all_comments?v={video_id}'
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    comments = []
    for comment in soup.find_all('yt-formatted-string', class_='style-scope ytd-comment-renderer'):
        comments.append(comment.text)
    return comments
```

```python
video_id = 'rxzhWUWV9yM'
comments = crawl_comments(video_id)
for comment in comments:
    print(comment)
```

```python
import pandas as pd   
df = pd.DataFrame(comments, columns=['comment'])
df
```
