```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
#manampung data dalam array
tweet_list = []
```

```python
#mangambil data twitter dengan TwitterSearchScrapper dan meletakan dalam array
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('Cak Imin since:2018-01-01 until:2023-02-20').get_items()):
  if i>1500:
    break
  tweet_list.append([tweet.user.username,tweet.content,tweet.user.location])
```

```python
frame_tweet = pd.DataFrame(tweet_list,columns =['Username','Konten','Lokasi'])
frame_tweet
```

```python
frame_tweet.to_csv('crawling cak imin.csv',index=0,sep=",")
```
