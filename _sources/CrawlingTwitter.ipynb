## Crawling data Berita Online

```python
!pip install requests
!pip install beautifulsoup4
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

### Mengambil data dari detik.com

```python
pencarian = 'snbp'
cari = pencarian.replace(' ','+')
tglAwal = '18/01/2022'
tglAkhir = '18/02/2023'
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
}
berita_list = []
for halaman in range(1):
  url = f'https://www.detik.com/search/searchall?query={cari}&sortby=time&fromdatex={tglAwal}&todatex={tglAkhir}&page={halaman}'
  #mengambil data dari detik.com
  req = requests.get(url,header)
  soup = BeautifulSoup(req.text,'lxml')
  listberita = soup.find('div',class_='list media_rows list-berita')
  artikel = listberita.find_all('article')
  for x in artikel:
    url2 = x.find('a')['href']
    judul = x.find('a').find('h2').text
    #mengambil data dari setiap konten
    urlkonten = requests.get(url2,header)
    soupkonten = BeautifulSoup(urlkonten.text,'lxml')
    konten = soupkonten.find_all('div',class_='detail__body-text itp_bodycontent')
    for x in konten:
      isi = x.find_all('p')
      y = [y.text for y in isi]
      fixkonten = ''.join(y).replace('\n','').replace('ADVERTISEMENT','').replace('SCROLL TO RESUME CONTENT','')
      berita_list.append([judul,fixkonten])
```

### Mengubah ke Data Frame

```python
frame_berita = pd.DataFrame(berita_list,columns =['Judul','Isi'])
frame_berita
```

### Import data ke csv

```python
frame_berita.to_csv('databerita.csv',sep=',',index=False)
```

## Crawling Website PTA

### Mengambil data dari web Pta Trunojoyp

```python
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
}
#Teknik_informatika = 10
pta_list = []
for halaman_prodi in range(1,5):
  for halaman in range(1,2):
    url = f"https://pta.trunojoyo.ac.id/c_search/byprod/{halaman_prodi}/{str(halaman+1)}"
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'lxml')
    prodi = soup.find('h2').text.replace('Journal','')
    list_ = soup.find('div', {'id':'content_journal'})

    for pta in list_.findAll('li',{'data-cat':'#luxury'}):
      Nama = pta.span.text.strip('\n').replace('Penulis : ', '')
      link = pta.find('a', 'gray button')['href']
      req = requests.get(link)
      soup2 = BeautifulSoup(req.text, 'lxml')
      judul = pta.a.text.strip('\n').replace('\r','').replace('\n','')
      abstrak = soup2.find('p', {'align':'justify'}).text.replace('ABSTRAK','').replace('\r','').replace('\n','').replace('\t','')
    
      NIM = link.replace('https://pta.trunojoyo.ac.id/welcome/detail/', '')

      pta_list.append([ NIM,Nama,judul,abstrak,prodi])
```

### Mengubah Ke Data Frame

```python
frame_pta = pd.DataFrame(pta_list, columns=['NIM','Nama','Judul','Abstrak','Prodi'])
frame_pta
```

### Import data ke csv

```python
frame_pta.to_csv('datapta.csv',sep=',',index=False)
```
